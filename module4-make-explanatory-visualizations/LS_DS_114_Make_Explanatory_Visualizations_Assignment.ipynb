## ASSIGNMENT PART 1: REPLICATE

```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a panda series that approximate the height of the bars in the graphic
fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33],
                 index=range(1,11))

# Set Stylesheet
plt.style.use('fivethirtyeight')

# figure / axes syntax
fig, ax = plt.subplots()

# plot the bars
fake.plot.bar(fake, width=0.9, color='#ED7139')

# title
ax.set_title("An Inconvenient Sequel: Truth to Power' is divisive",
             loc='left',
             color='black',
             fontsize=13,
             fontweight='bold',
             x=-0.07,
             y=1.1)

# Add Subtitle by adding a text object
ax.text(s='IMDb ratings for the film as of Aug. 29',
        x=-1.4,
        y=42,
        color='black',
        fontsize=12)

# X label
ax.set_xlabel('Rating', color='Black', fontsize=12, fontweight='bold')

# Y label
ax.set_ylabel('Percent of total values', color='Black', fontsize=12, fontweight='bold')

# X axis tick labels
ax.set_xticklabels(labels=range(1,11,1), color='gray', rotation='horizontal', fontsize='11')

# Y axis tick labels
ax.set_yticklabels(labels=['0', '10', '20', '30', '40%'], color='gray', fontsize='11')

# Set Y ticks (location of ticks)
ax.set_yticks(ticks=range(0,50,10))

# Set figure background color
fig.set_facecolor(color='white')

# Set the axes background color
ax.set_facecolor(color='white')
```

## ASSIGNMENT PART 2: REPRODUCE

### [Original Chart](https://fivethirtyeight.com/wp-content/uploads/2015/04/barry-jester-datalab-boomersdruguse-actual.png?w=575) 

```python
# Import Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set series for approximate height of bars
fake = pd.Series([7.29, 2.52, 1.43, 0.87, 0.36, 0.36,
                  0.28, 0.25, 0.23, 0.15, 0.15, 0.05],
                 index=range(1,13,1))

# Create Series for Y Tick Labels
drugs = ['Marijuana', 'Pain reliever', 'Tranquilizer', 'Cocaine', 'Crack', 
         'OxyContin', 'Stimulant', 'Hallucinogen', 'Sedative', 'Inhalant', 
         'Meth', 'Heroin']

# Set Plot Style
plt.style.use('fivethirtyeight')

# Set axes and figures
fig, ax = plt.subplots()

# Plot the bars
fake.plot.barh(fake, width=0.92, color='#FF2600')

ax.set_title('''Percentage of Americans aged 50-64 who said in a 2012 survey
that they had used the following drugs in the past year''', 
             loc='left',
             color='#3F3F3F',
             fontsize=15,
             x = -.25,
             y = 1.08)

# Set Y Tick Labels
ax.set_yticklabels(drugs, color='#3F3F3F', rotation='horizontal', fontsize=14)

# Set Y Labels top-to-bottom
ax.invert_yaxis()

# Make X-Axis Labels Invisible
ax.set_xticklabels([])

# Set background of chart to white
fig.set_facecolor('white')
ax.set_facecolor('white')

# Hide grid
plt.grid(False)

# Set Right Orient Y-Labels
for i in ax.patches: 
    plt.text(i.get_width()+0.2, i.get_y()+0.5,  
             str(round((i.get_width()), 2)), 
             fontsize = 11, 
             color ='#3F3F3F') 
```
